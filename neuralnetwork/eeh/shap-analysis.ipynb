## Load Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import shap
import seaborn as sns
import os
os.chdir('..')
import torch
import numpy as np
from src.model import predict
```

```python
# set plotting style
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['font.size'] = 14
plt.rcParams['figure.dpi'] = 200
fig_size = (8, 4)
```

## Load Experimental Data

```python
data = pd.read_csv('../data/cleaned_data.csv')
data.drop(columns=['S/N', 'C2H4', 'CO', 'H2', 'EtoH', 'FORM'], inplace=True)
data.head()
```

```python
output = pd.read_csv('../data/cleaned_data.csv')
output.drop(columns=['S/N'], inplace=True)
output.drop(columns=(list(data.columns)), inplace=True)
output.head()   
```

```python
# normalize the data with max values
data['Cu %'] /= 1.00
data['weight'] /= 118.71
data['Pot'] /= 4.70
data['pH'] /= 14.05
data['cDen'] /= 450.00
```

```python
data.head()
```

```python
# rerrange the columns
data = data[['Sn %', 'Pot', 'weight', 'pH', 'Cu %', 'cDen']]
data.head()
```

## Perform Shap Analysis

### 1. C2H4

```python
def f_instance(df):
    input_df = np.asarray(df)
    input = torch.from_numpy(input_df).float()
    out = predict(input, layer_model=[6, 20, 20, 15, 3], dir='./eeh/')[:, 0]

    return out

explainer = shap.Explainer(f_instance, data)
shap_values = explainer(data)
```

```python
fig1x = plt.figure()

ax1x0 = fig1x.add_subplot(121)
shap.summary_plot(shap_values, features=data, show=False,)
ax1x0.set_xlabel(r'$\rm Feature\ Importance \ (SHAP \ value)$')
ax1x0.set_title(r'$\rm (a)$')

ax1x1 = fig1x.add_subplot(122)
shap.summary_plot(shap_values, features=data, show=False, plot_type='bar')
ax1x1.set_xlabel(r'$\rm Feature\ Importance \ (SHAP \ value)$')
ax1x1.set_title(r'$\rm (b)$')

plt.gcf().set_size_inches(fig_size)
plt.tight_layout()
plt.show()
```

```python
idx_max, idx_min = np.argmax(output['C2H4']), np.argmin(output['C2H4'])
print(f"Min: {output['C2H4'].iloc[idx_min]*100}%, Max: {output['C2H4'].iloc[idx_max]*100}%")
```

```python
fig = plt.figure()

ax0 = fig.add_subplot(121)
shap.plots.waterfall(shap_values[idx_min], show=False)
ax0.set_title(r'$\rm Min \ C_2H_4$')

ax1 = fig.add_subplot(122)
shap.plots.waterfall(shap_values[idx_max], show=False)
ax1.set_title(r'$\rm Max \ C_2H_4$')

plt.gcf().set_size_inches(fig_size)
plt.tight_layout()
plt.show()
```

The SHAP (SHapley Additive exPlanations) values observed in this multi-output regression model indicate the contribution of each feature to the predicted outputs. Below is a summary of the observations:

The contributions of the features to the predictions of output 1 are quite minimal, indicating either the features are not very influential or the model is not sensitive to the input features for this particular output.

2. **Output 2(C2H5OH FE)**: SHAP values range from -0.02 to 0.01

3. **Output 3(H2 FE)**: SHAP values range from -0.05 to 0.25
   - The significantly wider range, especially with a positive upper bound as high as 0.25, suggests that the features have a substantial impact on this output. This indicates that the model's predictions for output 3 are highly influenced by the input features, with some features contributing positively and others negatively to a significant extent.

- **Feature Importance**: For output 3, identifying which features have the highest SHAP values could provide insights into the most important features driving the predictions.
- **Model Performance**: The varying ranges in SHAP values might reflect the model's performance across different outputs. If output 3 has the highest SHAP values, it might also have better predictive accuracy if those feature contributions align well with the actual outcomes.
