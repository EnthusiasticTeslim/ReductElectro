```python
import numpy as np

def equation_1(Pot, cDen):
    a0 = 6.249472314478992e-02
    c0 = 1.383660339372293e-01
    return c0 + a0 * ((Pot - cDen) / (Pot**6))

def equation_2(Cu, weight, Pot, cDen):
    a0 = -2.581083511586026e-01
    a1 = -8.664956150583759e-02
    c0 = 2.098383914436108e-01
    return c0 + a0 * (np.abs((Cu**6) - np.abs(weight - cDen))) + a1 * (np.log(cDen) / (Pot**3))

def equation_3(Cu, weight, Pot, cDen, pH):
    a0 = 1.125325009922978e-01
    a1 = -5.341844159351353e-01
    a2 = 7.855348236927703e-02
    c0 = 2.630605460651455e-01
    return c0 + a0 * (np.abs((Cu / cDen) - (1.0 / weight))) + a1 * (np.abs(np.exp(-1.0 * Pot) - (Cu**2))) + a2 * ((pH - cDen) / (weight * cDen))

def equation_4(cDen, pH, Pot, Sn, Cu, weight):
    a0 = -2.740000236048520e-01
    a1 = -3.111359138553260e-01
    a2 = -2.917950633181211e-01
    a3 = 8.147155720978794e-02
    c0 = 3.764992147354106e-01
    return c0 + a0 * (np.abs((cDen / pH) - np.cos(Pot))) + a1 * (np.abs(np.abs(Sn - Pot) - np.exp(-1.0 * weight))) + a2 * (np.abs((Cu**6) - np.abs(weight - cDen))) + a3 * (np.abs(Pot - cDen) / (Pot**6))

def equation_5(pH, cDen, Pot, Sn, Cu, weight):
    a0 = -1.327086673157861e-01
    a1 = -1.839947831565793e+00
    a2 = -2.411112167669870e-01
    a3 = -2.890421984924531e-01
    a4 = 6.996071260660672e-02
    c0 = 3.874641618781436e-01
    return c0 + a0 * (np.abs((pH - cDen) - (Pot**6))) + a1 * ((Sn**6) * (Cu * cDen)) + a2 * (np.abs(np.abs(Sn - Pot) - np.exp(-1.0 * Pot))) + a3 * (np.abs((Cu**6) - np.abs(weight - cDen))) + a4 * (np.abs(Pot - cDen) / (Pot**6))

def equation_6(Sn, Cu, Pot, pH, cDen):
    a0 = 3.277493220571178e-01
    a1 = -5.598909495281141e-01
    a2 = 9.546056163094251e-01
    a3 = -2.895967612298517e+00
    a4 = 1.204578438049861e-01
    a5 = -3.447184433917704e-01
    c0 = 1.886998352450209e-01
    return c0 + a0 * (np.abs(np.sqrt(Sn) - np.abs(Cu - Pot))) + a1 * (np.abs((pH - cDen) - (Pot**6))) + a2 * (np.abs((cDen / pH) - np.cos(Pot))) + a3 * ((Sn**6) * (Cu * cDen)) + a4 * (np.abs((Cu / cDen) - (Cu + Pot))) + a5 * (np.abs(np.sqrt(Sn) - (cDen / pH)))
```

```python
data['CO_pred_1'] = equation_1(data['Pot'], data['cDen'])
data['CO_pred_2'] = equation_2(data['Cu %'], data['weight'], data['Pot'], data['cDen'])
data['CO_pred_3'] = equation_3(data['Cu %'], data['weight'], data['Pot'], data['cDen'], data['pH'])
data['CO_pred_4'] = equation_4(data['cDen'], data['pH'], data['Pot'], data['Sn %'], data['Cu %'], data['weight'])
data['CO_pred_5'] = equation_5(data['pH'], data['cDen'], data['Pot'], data['Sn %'], data['Cu %'], data['weight'])
data['CO_pred_6'] = equation_6(data['Sn %'], data['Cu %'], data['Pot'], data['pH'], data['cDen'])
```

```python
def percentage_mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred)) * 100
```
